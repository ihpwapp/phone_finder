```python
# Create 100 variables with names test1, test2, ..., test100

# Create a list with 100 elements using a list comprehension
variable_list = [globals()[f"rule{i}"] for i in range(1, 82)]
print(len(variable_list))
```

```python
#creating an ouptut value
landslideCtrl = ctrl.ControlSystem(variable_list)
landSlideEval = ctrl.ControlSystemSimulation(landslideCtrl)
```

```python
customtkinter.set_appearance_mode("System")  # Modes: system (default), light, dark
customtkinter.set_default_color_theme("blue")  # Themes: blue (default), dark-blue, green

app = customtkinter.CTk()  # create CTk window like you do with the Tk window
app.geometry("420x240")
app.title("Landslide Occurence Odds")

def landslideCal(r,l,v,p):
    landSlideEval.input['rainfall'] = r
    landSlideEval.input['landSlope'] = l
    landSlideEval.input['vibration'] = v
    landSlideEval.input['plasticityIndex'] = p
    landSlideEval.compute()
    landslide.view(sim=landSlideEval)
    return landSlideEval.output['landslide']

def button_function():
    i1 = int(entry1.get())
    i2 = int(entry2.get())
    i3 = int(entry3.get())
    i4 = int(entry4.get())
    output1 = landslideCal(i1,i2,i3,i4)
    print(output1)
    
def buttonInput():
    dialog = customtkinter.CTkInputDialog(text="Type in a number",title="test")
    print("Number",dialog.get_input())
```

```python
text = customtkinter.CTkLabel(app, text="Landslide Input")
button = customtkinter.CTkButton(master=app, text="Calculate", command=button_function)
button.grid(row=3,column=1,padx=20)
entry1 = customtkinter.CTkEntry(app, placeholder_text="Rainfall")
entry2 = customtkinter.CTkEntry(app, placeholder_text="landslope")
entry3 = customtkinter.CTkEntry(app, placeholder_text="vibration")
entry4 = customtkinter.CTkEntry(app, placeholder_text="Plasticicty Index")


text.grid(row =0 ,columnspan=3,sticky = "n")
entry1.grid(row=1,column=0,padx=(20,0))
entry2.grid(row=2,column=0,padx=(20,0))
entry3.grid(row=3,column=0,padx=(20,0))
entry4.grid(row=4,column=0,padx=(20,0))


app.mainloop()
```
